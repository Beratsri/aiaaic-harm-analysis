# 03 - Harm Classifiers (Individual and Societal)
This notebook trains TF-IDF + Logistic Regression classifiers to predict missing `Harm_Individual` and `Harm_Societal` labels.

```python
import pandas as pd
import sys
import os

sys.path.append("../src")
from config import CLEAN_DATA_PATH, HARM_INDIVIDUAL_MODEL_PATH, HARM_SOCIETAL_MODEL_PATH
from models.harm_classifier import HarmClassifier
```

## Load cleaned data

```python
df_clean = pd.read_csv(CLEAN_DATA_PATH)
print(f"Loaded clean data: {df_clean.shape}")
```

## Train Harm_Individual Model

```python
print("Training Harm_Individual Classifier...")
clf_ind = HarmClassifier(target_column='Harm_Individual', top_n_classes=15)
clf_ind.fit(df_clean)
clf_ind.save(HARM_INDIVIDUAL_MODEL_PATH)
```

## Train Harm_Societal Model

```python
print("Training Harm_Societal Classifier...")
clf_soc = HarmClassifier(target_column='Harm_Societal', top_n_classes=15)
clf_soc.fit(df_clean)
clf_soc.save(HARM_SOCIETAL_MODEL_PATH)
```
