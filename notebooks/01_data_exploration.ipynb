# 01 - Exploratory Data Analysis
This notebook explores the raw AIAAIC dataset to understand missing values, yearly trends, developers, technologies, and harm categories.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import os
import sys

sys.path.append("../src")
from config import RAW_DATA_PATH, ensure_dirs
ensure_dirs()
sns.set_theme(style="whitegrid")
```

## Load Raw Excel Data

```python
df_raw = pd.read_excel(RAW_DATA_PATH, sheet_name='Incidents', header=[1, 2])
print(f"Raw data shape: {df_raw.shape}")
df_raw.head(3)
```

## Missing Value Analysis

```python
# Delineate multi-index columns for plotting
flat_cols = [f"{c[0]} - {c[1]}" if "Unnamed" not in str(c[1]) else c[0] for c in df_raw.columns]
df_temp = df_raw.copy()
df_temp.columns = flat_cols

plt.figure(figsize=(12, 6))
msno.matrix(df_temp)
plt.title("AIAAIC Dataset Missing Values Matrix", fontsize=16)
plt.show()
```

## Raw Column Missingness Percentages

```python
missing_pct = (df_raw.isna().sum() / len(df_raw)) * 100
missing_df = pd.DataFrame({"Column": flat_cols, "Missing %": missing_pct.values})
missing_df = missing_df.sort_values("Missing %", ascending=False)
missing_df
```
